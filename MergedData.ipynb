```python
import pandas as pd

# List of years
years = [2020, 2021, 2022, 2023, 2024]

# List to hold dataframes
df_list = []

# Load and process each year's data
for year in years:
    # Assume files are named like 'brfss_{year}.csv' in current directory
    file_path = f'brfss_{year}.csv'  
    
    # Load (use pd.read_parquet if Parquet files)
    df_year = pd.read_csv(file_path, low_memory=False)
    
    # Add Year column
    df_year['Year'] = year
    
    df_list.append(df_year)

# Merge all years
merged_df = pd.concat(df_list, ignore_index=True)

# Save as Parquet
merged_df.to_parquet('merged_brfss_diabetes_key_2020_2024.parquet', index=False)

print("Merged file saved as 'merged_brfss_diabetes_key_2020_2024.parquet'")
```
